```python
#1(a)
def factorial(n):
  if n == 0:
    return 1
  a = n * factorial(n-1)
  return a
factorial(3)
```

```python
#1(b)
def fibonacci(n):


  if n==0:
    return 0
  elif n == 1:
    return 1
  a = fibonacci(n-1)+fibonacci(n-2)
  return a

fibonacci(5)
```

```python
#1(d)
def powerN(b,p):
  if p == 0 :
    return 1
  a = b*powerN(b,p-1)
  return a
print(powerN(3,3))
```

```python
#2(a)
def decimal_to_binary(n):

  if n == 1:
    e=str(1)
    return e

  e=  decimal_to_binary(n//2) + str(n%2)


  return e
decimal_to_binary(9)
```

```python
#2(b)
class Node:
  def __init__(self,e,n=None):
    self.elem = e
    self.next = n
#to visualize the linked list
def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()

def createList(arr):
  head = Node(arr[0])
  temp = head
  for i in range(1,len(arr)):
    new = Node(arr[i])
    temp.next = new
    temp = temp.next
  return head
def addList(head):
  if head == None:
    return 0
  r = head.elem + addList(head.next)
  return r
l = createList([431,2,3,432,5])
printLinkedList(l)
print(addList(l))

```

```python
#2(c)
def reverseList(head):
  if head == None:
    return
  reverseList(head.next)
  print(head.elem)
l = createList([431,2,3,432,5])
printLinkedList(l)
reverseList(l)

```

```python
#3
def hocBuilder(height):
  if height == 0 :
    return 0
  if height == 1 :
    return 8
  r = 5 + hocBuilder(height-1)
  return r
hocBuilder(5)
```

```python
#4(b)
def num(n):
  if n == 1:
    print(1,end='')
    return
  num(n-1)
  print(n,end='')
def space(n):

  print(' '*(n),end='')


def pattern2(n,s=0):
  if n == 1:
    space(s)
    num(1)
    print()
    return
  pattern2(n-1,s+1)
  space(s)
  num(n)
  print()
pattern2(7)

```
